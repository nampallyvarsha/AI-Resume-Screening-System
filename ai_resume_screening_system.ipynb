```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
Create data set
```

```python
data = {
"resume":[
"python machine learning data analysis",
"deep learning python artificial intelligence",
"html css javascript web development",
"digital marketing seo content marketing",
"sql excel data visualization analyst",
"react javascript frontend development"
],

"role":[
"Data Scientist",
"Data Scientist",
"Web Developer",
"Marketing",
"Data Analyst",
"Web Developer"
]
}
```

```python
Convert data frame
```

```python
df = pd.DataFrame(data)
df
```

```python
TF-IDE vectorization
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["resume"])
y = df["role"]
```

```python
model = LogisticRegression()
model.fit(X,y)
```

```python
Train model
```

```python
test_resume = ["python sql data visualization"]

test_vector = vectorizer.transform(test_resume)

prediction = model.predict(test_vector)

print("Predicted Job Role:", prediction[0])
```

```python
test_resume = ["html css javascript website"]

test_vector = vectorizer.transform(test_resume)

prediction = model.predict(test_vector)

print("Predicted Job Role:", prediction[0])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print("Model Accuracy:", accuracy)
```
